```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
file_path = 'NewAQI.csv'  # Replace with the actual path to your CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe for verification
print(df.head())
```

```python
# Step 2: Bar Graph - Comparison of AQI values for different places within each country
plt.figure(figsize=(12, 8))

for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    plt.bar(country_data['City'], country_data['AQI Value'], label=country)

plt.title('Comparison of AQI Values Across Different Places Within Each Country')
plt.xlabel('City')
plt.ylabel('AQI Value')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Step 3: Pie Chart - Distribution of AQI categories across all places
plt.figure(figsize=(8, 8))

category_counts = df['AQI Category'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral', 'lightblue'])
plt.title('Distribution of AQI Categories Across All Places')
plt.show()
```
